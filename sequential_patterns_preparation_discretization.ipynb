# Discretizzazione II: SAX

#### "SAX is the first symbolic representation for time series that allows for dimensionality reduction and indexing with a lower-bounding distance measure." (http://www.cs.ucr.edu/~eamonn/SAX.htm)

#### "In short, Symbolic Aggregate approXimation (SAX) algorithm application to the input time series transforms its into a strings. The algorithm was proposed by Lin et al.) and extends the PAA-based approach inheriting the original algorithm simplicity and low computational complexity while providing satisfactory sensitivity and selectivity in range query processing. Moreover, the use of a symbolic representation opened a door to the existing wealth of data-structures and string-manipulation algorithms in computer science such as hashing, regular expression, pattern matching, suffix trees, and grammatical inference."

#### PAA: "A time series C of length n can be represented in a w-dimensional space.Simply  stated,  to  reduce  the  time  series from n dimensions  to  w dimensions, the data is divided into w equal sized frames. The mean  value  of  the  data  falling  within  a  frame  is calculated  and  a  vector  of  these  values becomes  the  data-reduced  representation.  The representation can be visualized as an attempt to approximate the  original  time  series  with  a  linear  combination  of  box  basis  functions." (https://cs.gmu.edu/~jessica/Lin_motif.pdf)

#### Discretization: "Having  transformed  a  time  series  database  into  the  PAA  we  can  apply  a  further  transformation  to  obtain  a discrete  representation.  For  reasons  that  will  become  apparent  in  Section  4,  we  require  a  discretization    technique    that    will   produce    symbols    with    equiprobability.  This  is  easily  achieved  since  normalized  time  series have a Gaussian distribution." (https://cs.gmu.edu/~jessica/Lin_motif.pdf) "There is one issue we must address if we are to use a symbolic representation of time series. If we wish to approximate a massive dataset in main memory, the parameters w and a have to be chosen in such a way that the approximation makes the best use of the primary memory available. There is a clear tradeoff between the  parameter  w  controlling  the  number  of  approximating  elements,  and  the  value  a  controlling  the  granularity of each approximating element. (...) The  results  suggest  that using  a  low  value  for  a  results  in  weak  bounds.  While  its  intuitive  that  larger  alphabet  sizes  yield  better  results,  there  are  diminishing  returns  as  a increases.  If  space  is  an  issue,  an  alphabet  size  in  the  range  5  to  8  seems  to  be  a  good  choice  that  offers  a  reasonable  balance  between  space  and tightness of lower bound  each alphabet within this range can be represented with just 3 bits. Increasing the alphabet size would require more bits to represent each alphabet." (https://cs.gmu.edu/~jessica/SAX_DAMI_preprint.pdf)


```python
from saxpy.znorm import znorm
from saxpy.sax import ts_to_string 
from saxpy.alphabet import cuts_for_asize
from saxpy.paa import paa
```

## Generazione dei file di input per SPFM

#### Viene generato un file di input diverso per ogni alfabeto provato (5,6,7, o 8 lettere). Il file deve avere le stesse caratteristiche viste in precedenza. Nel nostro caso ogni serie dell'insieme {(1,1962),(2,1962),(3,1962),...} corrisponde a una sequenza costituita da un numero variabile di itemsets (i giorni) ognuno di un solo elemento (l'intero che rappresenta il bin di appartenenza del valore di Open per ogni giorno). Affinch ogni giorno sia rappresentato da un intero  necessario convertire prima le serie letterali in serie numeriche.

```python
#Funzione generica per la generazione dei file di input per SPFM
def genera_input (alphabet, dictionary, file):
    dictionary_n=OrderedDict() #dizionario in cui inserire le serie discretizzate convertite in serie numeriche
    for key, serie in dictionary.items():
        serie_n=list()
        for item in serie: 
            serie_n.append(alphabet[item])
        dictionary_n[key]=serie_n
    #Generazione del file di input
    for key, serie in dictionary_n.items():
        for item in serie:
            file.write(item)
            file.write(" ")
            file.write("-1 ")
        file.write("-2\n")
    return "Generato file di input"
```

```python
#Generazione file di input per alfabeto=5
alphabet5={'a':'1','b':'2','c':'3','d':'4','e':'5'}
input_file=open('input_file_sax5.txt', 'w')
print(genera_input (alphabet5, groups_dict_sax_binned5, input_file))
input_file.close()
```
